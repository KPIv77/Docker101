```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generating data for plotting
x = np.linspace(0, 10, 100)  
y = np.sin(x)                
```

```python
# Plotting the sine wave
plt.plot(x, y)
plt.title("Simple Sine Wave")
plt.xlabel("X")
plt.ylabel("sin(X)")
plt.grid(True)
plt.show()
```
